```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
```

```python
def f(x, y):
    return x ** 3 - 3 * x * y ** 2

def df(x, y):
    return np.array([
        3 * x ** 2 - 3 * y ** 2,
        - 6 * x * y,
    ])
```

```python
x = np.linspace(-1., 1., 100)
y = np.linspace(-1., 1., 100)

X, Y = np.meshgrid(x, y)
plt.figure(figsize=(10, 10))
plt.contour(X, Y, f(X, Y), 10, cmap=plt.cm.RdBu)
plt.contourf(X, Y, f(X, Y), 100, cmap=plt.cm.RdBu);
```

```python
def eval_params(x):
    return f(x[0], x[1]), df(x[0], x[1])

eval_params([0., 0.])
```

```python
from scipy.optimize import fmin_l_bfgs_b

fmin_l_bfgs_b(eval_params, np.array([1, 0.0001]), disp=10)
```

```python
fmin_l_bfgs_b(eval_params, np.array([-.002, .003]), disp=10)
```

```python
x = np.linspace(-.01, .01, 101)
y = np.linspace(-.01, .01, 101)
pairs = np.dstack(np.meshgrid(x, y)).reshape(-1, 2)

attracted = []
final = []
stuck = []

all_results = []

for x, y in pairs:
    path = [(x, y)]
    cb = lambda x: path.append(x.copy())
    (x_final, y_final), obj_final, info = fmin_l_bfgs_b(eval_params, [x, y], pgtol=1e-4, callback=cb)
    converged = info['task'].startswith(b'CONVERGENCE: ') and np.abs(obj_final) < 0.1
    if converged:
        if len(path) > 1:
            attracted.append((x, y))
            final.append((x_final, y_final))
        else:
            stuck.append((x, y))
        #print("Path with %d elements:" % len(path))
    all_results.append(dict(path=path, info=info, obj_final=obj_final))
            
attracted = np.array(attracted)
final = np.array(final)
stuck = np.array(stuck)
```

```python
radius = pairs.max() - pairs.min()
x = np.linspace(pairs[:, 0].min() - radius / 10, pairs[:, 0].max() + radius / 10, 101)
y = np.linspace(pairs[:, 1].min() - radius / 10, pairs[:, 1].max() + radius / 10, 101)

X, Y = np.meshgrid(x, y)
plt.figure(figsize=(12, 12))
plt.contour(X, Y, f(X, Y), 10, cmap=plt.cm.RdBu)
plt.contourf(X, Y, f(X, Y), 100, cmap=plt.cm.RdBu)
plt.scatter(pairs[:, 0], pairs[:, 1], c='b', alpha=0.1)
plt.scatter(stuck[:, 0], stuck[:, 1], c='g')
if attracted.size:
    plt.scatter(attracted[:, 0], attracted[:, 1], c='b')
if final.size:
    plt.scatter(final[:, 0], final[:, 1], c='r');
```

```python
stuck = [r for r in all_results if np.abs(r['obj_final']) < 1]
```
